```python
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# User query
query = "How can I reset my password if I forgot it?"

# Knowledge base documents
documents = [
    "Recovering access to your account is simple. Visit the login page, select 'Forgot Credentials,' and follow the steps.",
    "Learn how to update your email address in your profile settings.",
    "To reset your login details, use the password recovery option available on the sign-in screen.",
    "Tips for creating a strong password to secure your account."
]

# Compute embeddings
query_embedding = model.encode(query, convert_to_tensor=True)
document_embeddings = model.encode(documents, convert_to_tensor=True)

# Compute cosine similarities
similarities = util.pytorch_cos_sim(query_embedding, document_embeddings)

# Rank documents by similarity
sorted_indices = similarities.argsort(descending=True)
for idx in sorted_indices[0]:
    print(f"Document: {documents[idx]}, Similarity: {similarities[0][idx].item()}")
```
